```
first_num = 1
while first_num <= 10:
  second_num = 0
  while second_num <= 10:
    multiply = first_num * second_num
    print(first_num, " * ", second_num, " = ", multiply, "\n")
    second_num += 1
  first_num += 1
  print("\n")
  
```
