```
#Our input will be a matrix to demonstrate how for loops work.

mat = []
base = 2
lbs = 0

for row in range(4):
  row = np.random.randint(10, size = 7).tolist()
  mat.append(row)

print(mat)
```

Turn each row of the matrix into the binary matrix. 
Use 3 different for loops: use an element vise iteration, use range and finally use enumerate. 

```
#1. Use iterators 

binrow = [0]*4
num = 0
bincol =[[0]*4]*6

matbinrow = []

for i in range(0,len(mat[0])):
  for j in range(0,4):
    lbs = mat[0][i] % base
    binrow[4-j-1] = lbs
    mat[0][i]//= base
  print(binrow) 
bincol.append(binrow)      
print(bincol)  
```
