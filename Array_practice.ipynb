```
def sumOddLengthSubarrays(self, arr):
  sum = 0
  for length in range(1,len(arr)+1,2):        
    for start in range(len(arr)):  
      if(start +length <= len(arr)):                    
        for index in range(start,start+length):
          sum = sum + arr[index]
                   
  return sum
```
